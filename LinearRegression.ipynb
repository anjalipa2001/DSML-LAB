```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
#load the datasets
df=pd.read_csv("/content/insurance_data.csv")
df.head()
df.isnull().sum()
```

```python
#Regression plot between age and charges
import seaborn as snb
import matplotlib.pyplot as plt
snb.regplot(x=df['age'],y=df['charges'],data=df)
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
#Splitting the datasets
x=df.iloc[:,0]
y=df.iloc[:,-1]
print(x)
print(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=3)
```

```python
#into 2D array
x_train.shape
x_test.shape
y_train.shape
y_test.shape
x_train_re=x_train.values.reshape(-1,1)
x_test_re=x_test.values.reshape(-1,1)
y_train_re=y_train.values.reshape(-1,1)
y_test_re=y_test.values.reshape(-1,1)
```

```python
#model
lr=LinearRegression()
lr.fit(x_train_re,y_train_re)
print("slope: ",lr.coef_)
print("Intercept : ",lr.intercept_)
```

```python
#predict the model for test and train
y_pred_test=lr.predict(x_test_re)
y_pred_train=lr.predict(x_train_re)
print(y_pred_test)
y_pred_test.shape
```

```python
#Evaluating the model
print("R2 score : ",r2_score(y_test_re,y_pred_test))
print("Mean Squared Error : ",mean_squared_error(y_test_re,y_pred_test))
print("Mean Absolute error : ",mean_absolute_error(y_test_re,y_pred_test))

```

```python
#plotting using scatter
plt.scatter(x_train_re,y_train_re,color='blue',label='Training dats')
plt.plot(x_train_re,y_pred_train,color='red',label='Regression line')
plt.title('LINEAR REGRESSION')
plt.xlabel('age')
plt.ylabel('charges')
plt.legend()
plt.show()
```
